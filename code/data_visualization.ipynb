```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clonotype = pd.read_csv('BCR_UMI.txt', sep='\t')
```

```python
clonotype
```

```python
# plot
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(13.7, 10.27)
sns.boxplot(data=clonotype[['M8000_R1', 'M8000_R2', 'M8000_R3', 'M8000_R4', 'M4000_R1', 'M4000_R2',
       'M4000_R3', 'M4000_R4']], palette=['red','red', 'red', 'red', 'green', 'green', 'green','green'])    
sns.despine()


plt.xlabel('Theoretical spike-in molecules', fontsize=16);
plt.ylabel('UMI count of each BCR clonotypes', fontsize=16);
plt.title('48 BCR clonotypes UMI count distribution', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xticks(rotation=270)
plt.ylim(0, 2500)
fig.savefig('BCR.png')
```
